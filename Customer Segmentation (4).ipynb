# Task
Remove outliers from the dataframe using the IQR method.

## Calculate iqr for numerical columns

### Subtask:
Compute the first quartile (Q1), third quartile (Q3), and Interquartile Range (IQR) for each numerical column.


**Reasoning**:
Calculate Q1, Q3, and IQR for each numerical column to prepare for outlier removal.



```python
numerical_cols = df.select_dtypes(include=np.number)
Q1 = numerical_cols.quantile(0.25)
Q3 = numerical_cols.quantile(0.75)
IQR = Q3 - Q1
```

## Define outlier bounds

### Subtask:
Determine the lower and upper bounds for outlier detection using the IQR method.


**Reasoning**:
Calculate the lower and upper bounds for outlier detection using the IQR method.



```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound for Outliers:\n", lower_bound)
print("\nUpper Bound for Outliers:\n", upper_bound)
```

## Remove outliers

### Subtask:
Filter the DataFrame to exclude rows where values in numerical columns fall outside the defined bounds.


**Reasoning**:
Create a boolean mask to identify rows with outliers in any numerical column and then filter the DataFrame to exclude those rows.



```python
outlier_mask = (numerical_cols < lower_bound) | (numerical_cols > upper_bound)
rows_to_remove_mask = outlier_mask.any(axis=1)
df = df[~rows_to_remove_mask].copy()
display(df.head())
```

## Verify outlier removal

### Subtask:
Display the shape of the DataFrame and generate boxplots to confirm that outliers have been removed.


**Reasoning**:
Display the shape of the filtered DataFrame and generate boxplots for numerical columns to visualize the removal of outliers.



```python
print("Shape of DataFrame after outlier removal:", df.shape)

num_cols = df.select_dtypes(include=np.number).columns.tolist()

for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col} after outlier removal')
    plt.show()
```

# **Model building**

```python
# Step 4: Select features for clustering
# Choose numeric features only
features = df.select_dtypes(include=np.number)
```

```python
# Step 5: Standardize features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
```

```python
# Step 6: Elbow Method to find optimal K
from sklearn.cluster import KMeans
wcss = []
K_range = range(2, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS (Inertia)")
plt.title("Elbow Method for Optimal K")
plt.show()
```

```python
# Step 7: Silhouette Score Evaluation
from sklearn.metrics import silhouette_score
silhouette_scores = []
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)

plt.figure(figsize=(8, 5))
plt.plot(K_range, silhouette_scores, marker='o', color='green')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score vs Number of Clusters")
plt.show()
```

```python
# Step 8: Fit KMeans with chosen K (example: K=4)
optimal_k = 4  # Change based on elbow + silhouette results
final_model = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = final_model.fit_predict(X_scaled)
```

```python
# Calculate silhouette score for the chosen k
score = silhouette_score(X_scaled, df['Cluster'])
print(f"Silhouette Score for K={optimal_k}: {score}")
```

```python
# Step 9: Cluster Summary
cluster_summary = df.groupby('Cluster').mean(numeric_only=True)
print("\n📊 Cluster Summary (mean values):\n", cluster_summary)
```

```python
# Step 10: Visualize Clusters (PCA 2D projection)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
reduced = pca.fit_transform(X_scaled)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=reduced[:,0], y=reduced[:,1], hue=df['Cluster'], palette="Set2")
plt.title("Customer Clusters (PCA Projection)")
plt.show()
```

```python
# 🧩 Customer Segmentation App with Streamlit

import streamlit as st
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

# --------------------------------------
# Title and description
# --------------------------------------
st.title("📊 Customer Segmentation App")
st.write("Upload your marketing dataset and discover customer segments using KMeans clustering.")

# --------------------------------------
# File upload
# --------------------------------------
uploaded_file = st.file_uploader("Upload your Excel file", type=["xlsx", "csv"])

if uploaded_file is not None:
    # Read data
    if uploaded_file.name.endswith(".csv"):
        df = pd.read_csv(uploaded_file)
    else:
        df = pd.read_excel(uploaded_file)

    st.write("### Preview of Dataset")
    st.dataframe(df.head())

    # Basic cleaning
    if 'Year_Birth' in df.columns:
        df['Age'] = 2025 - df['Year_Birth']

    # Drop missing values
    df = df.dropna()

    # Select numeric features
    features = df.select_dtypes(include=np.number)
    for col in features.columns:
        if "ID" in col or "Id" in col:
            features = features.drop(columns=[col])

    # Scale
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(features)

    # Sidebar controls
    st.sidebar.header("⚙️ Clustering Settings")
    k = st.sidebar.slider("Select number of clusters (K)", 2, 10, 4)

    # Fit KMeans
    kmeans = KMeans(n_clusters=k, random_state=42)
    df['Cluster'] = kmeans.fit_predict(X_scaled)

    # Silhouette Score
    sil_score = silhouette_score(X_scaled, df['Cluster'])
    st.sidebar.write(f"📈 Silhouette Score: {sil_score:.3f}")

    # Cluster Summary
    st.write("### 📊 Cluster Summary (mean values)")
    st.dataframe(df.groupby("Cluster").mean())

    # PCA for visualization
    pca = PCA(n_components=2)
    reduced = pca.fit_transform(X_scaled)

    fig, ax = plt.subplots(figsize=(8, 6))
    sns.scatterplot(x=reduced[:, 0], y=reduced[:, 1], hue=df['Cluster'], palette="Set2", ax=ax)
    plt.title("Customer Clusters (PCA Projection)")
    st.pyplot(fig)

    # Download clustered dataset
    st.write("### 💾 Download Results")
    csv = df.to_csv(index=False).encode("utf-8")
    st.download_button("Download Clustered Dataset", csv, "clustered_customers.csv", "text/csv")
```
